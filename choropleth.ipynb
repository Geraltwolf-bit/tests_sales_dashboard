```python
import requests

gejson_url = 'https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/russia.geojson'

response = requests.get(gejson_url)

r_geojson = response.json()
```

```python
gdf = gpd.GeoDataFrame.from_features(r_geojson['features'])
gdf
```
